```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
sbn.set_style('whitegrid')
```

```python
x=np.linspace(1,30,30)
y1=np.random.randn(30)
y2=np.random.randint(3,40,30)
plt.figure(figsize=(12,3),dpi=90)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
fig,ax=plt.subplots(1,1,figsize=(12,3),dpi=90)
ax.plot(x,y1,label='left')
ax.set_ylim(-3,3)
ax.legend(loc='upper left')
#x
ax2=ax.twinx() 
ax2.plot(x,y2,color='red',label='right')
ax2.set_ylim(3,40)
ax2.legend(loc='upper right')
plt.tight_layout()
plt.grid()
plt.show()
```
